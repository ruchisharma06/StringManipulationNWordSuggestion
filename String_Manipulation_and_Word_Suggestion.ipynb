```python
import nltk

# Download the NLTK library data for wordnet
nltk.download('wordnet')

from nltk.corpus import wordnet
from itertools import product

def delete_char(word):
    """
    Given a word, returns all possible strings that have one character removed.
    """
    deletions = [word[:i] + word[i+1:] for i in range(len(word))]
    return deletions


word_to_edit = "pace"

deleted_words = delete_char(word_to_edit)


print("Deleted words:", deleted_words)

```

```python
from nltk.corpus import wordnet
from itertools import product


def switch_char(word):
    """
    Given a word, returns all possible strings that have two adjacent letters switched.
    """
    switches = [word[:i] + word[i+1] + word[i] + word[i+2:] for i in range(len(word)-1)]
    return switches



word_to_edit = "eta"


switched_words = switch_char(word_to_edit)



print("Switched words:", switched_words)
```

```python
len(switched_words)
```

```python
def replace_char(word):
    """
    Given a word, returns all possible strings that have one character replaced by another different character.
    """
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    replacements = [word[:i] + c + word[i+1:] for i, c in product(range(len(word)), alphabet) if c != word[i]]
    return replacements


word_to_edit = "can"


replaced_words = replace_char(word_to_edit)


print("Replaced words:", replaced_words)
```

```python
def insert_char(word):
    """
    Given a word, returns all possible strings that have an additional character inserted.
    """
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    insertions = [word[:i] + c + word[i:] for i, c in product(range(len(word)+1), alphabet)]
    return insertions


word_to_edit = "at"

inserted_words = insert_char(word_to_edit)

print("Inserted words:", inserted_words)

```

```python
print(len(inserted_words))
```

```python
def edit_distance_one(word, allow_switches=True):
    """
    Generate all potential edits that are a single modification away from a given word.
    """
    deletions = delete_char(word)
    replacements = replace_char(word)
    insertions = insert_char(word)

    if allow_switches:
        switches = switch_char(word)
        one_edit_set = set(deletions + replacements + insertions + switches)
    else:
        one_edit_set = set(deletions + replacements + insertions)

    return one_edit_set


word_to_edit = "At"

one_edit_words = edit_distance_one(word_to_edit)

print("One edit words:", one_edit_words)
```

```python
# Count the number of words in one_edit_words
num_one_edit_words = len(one_edit_words)

print("Number of one edit words:", num_one_edit_words)
```

```python
def edit_distance_two(word, allow_switches=True):
    """
    Generate all potential edits that are two modifications away from a given word.
    """
    one_edit_set = edit_distance_one(word, allow_switches)

    # For each word in the one-edit set, generate one edit
    two_edit_set = set()
    for edited_word in one_edit_set:
        two_edit_set.update(edit_distance_one(edited_word, allow_switches))

    return two_edit_set



word_to_edit = "At"

two_edit_words = edit_distance_two(word_to_edit)

print("Two edit words:", two_edit_words)


```

```python
# Count the number of words in two_edit_words
num_two_edit_words = len(two_edit_words)

print("Number of two edit words:", num_two_edit_words)
```

```python
from collections import Counter

# Assume you have already defined the edit_distance_one and edit_distance_two functions
# and you have read the vocabulary from the text file using Your_Read_Function

def Your_Read_Function(file_path):
    """
    function to read the vocabulary from a text file.
    """
    # Example implementation
    with open(file_path, 'r', encoding='utf-8') as file:
        vocabulary = set(file.read().split())
        #print(len(vocabulary))
    return vocabulary

def fix_edits(word, vocabulary, n=7):
    """
    Generate a list of zero to n potential suggestion tuples in the format (word, probability_of_word).
    """
    suggestions = []

    # Step 1: Generate suggestions based on the suggestion algorithm
    if word in vocabulary:
        suggestions.append((word, 1.0))
    else:
        one_edit_set = edit_distance_one(word)
        two_edit_set = edit_distance_two(word)

        possible_suggestions = vocabulary.intersection(one_edit_set) or vocabulary.intersection(two_edit_set)

        if possible_suggestions:
            total_distance = len(word)
            suggestions.extend([(suggestion, 1 / (total_distance + 1)) for suggestion in possible_suggestions])
        else:
            suggestions.append((word, 0.0))

    # Step 2: Formulate a 'best_words' dictionary with probabilities
    best_words = {suggestion: probability for suggestion, probability in suggestions}

    # Step 3: Choose the n best suggestions
    n_best_suggestions = Counter(best_words).most_common(n)

    return n_best_suggestions


file_path = '/content/shakespeare.txt'
vocabulary = Your_Read_Function(file_path)
word_to_fix = 'tot'
num_suggestions = 7

suggestions = fix_edits(word_to_fix, vocabulary, n=num_suggestions)

print(f"entered word = {word_to_fix}")
print(f"suggestions = {vocabulary.intersection(edit_distance_one(word_to_fix)).union(edit_distance_two(word_to_fix))}")
for i, (suggestion, probability) in enumerate(suggestions):
    print(f"word {i}: {suggestion}, probability {probability:.6f}")
```
